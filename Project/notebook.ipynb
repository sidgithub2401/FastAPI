```python
import pandas as pd 

df = pd.read_csv(r"C:\Users\Sid\Downloads\dummy_insurance_data.csv")
```

```python
df_copy = df 
df_copy
```

```python
df_copy["bmi"] = df["weight"] / (df["height"] **2)
```

```python
df_copy
```

```python
def age_group(age):
    if age<=18:
        return "young"
    elif age<45:
        return "adult"
    elif age<60:
        return "middle_aged"
    else:
        return "senior"
```

```python
df_copy["age_group"]= df_copy["age"].apply(age_group)
```

```python
def lifestyle_risk(row):
    if row["smoker"] == True and row["bmi"] > 30:
        return "High"
    elif row["smoker"] == True and row["bmi"] >27:
        return "Medium"
    else:
        return "low"
```

```python
df_copy["lifestyle_risk"] = df_copy.apply(lifestyle_risk,axis=1)
```

```python
df_copy
```

```python
tier1 = ['Mumbai','Pune','Banglore','Chennai']

def city_tier(city):
    if city in tier1:
        return 1
    else:
        return 2
    
df_copy['city'] = df_copy['city'].apply(city_tier) 
```

```python
df_copy
```

```python
df_copy.drop(columns=['age','weight','height','smoker'],inplace=True)
```

```python
df_copy
```

```python
X = df_copy[['income_lpa','city','occupation','bmi','age_group','lifestyle_risk']]
Y = df_copy[['insurance_premium_category']]
```

```python
X
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report , accuracy_score
categorical_features = ["age_group","occupation","lifestyle_risk"]
numerical_features = ['income_lpa','city','bmi']

preprocessor = ColumnTransformer(
    transformers=[
        ("cat",OneHotEncoder(),categorical_features),
        ("num","passthrough",numerical_features)
    ]
)
```

```python
pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier",RandomForestClassifier(random_state=42))
])
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=1)
pipeline.fit(X_train,Y_train)
```

```python
y_pred = pipeline.predict(X_test)
accuracy_score(Y_test,y_pred)
```

```python
import pickle
with open("model1.pkl","wb") as f:
    pickle.dump(pipeline,f)
```
